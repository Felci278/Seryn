```python
import pandas as pd
import numpy as np
```

```python
food_df = pd.read_csv('food_label.csv')
```

```python
def assign_gi_label(gi):
	if gi < 35:
		return 'friendly'
	elif 35 <= gi < 45:
		return 'limited'
	else:
		return 'Avoid'

food_df['GI_Label'] = food_df['Glycemic Index'].apply(assign_gi_label)
food_df.to_csv('food_labels_with_gi_label.csv', index=False)
```
